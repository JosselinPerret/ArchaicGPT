# Tokenize our data

Create a mapping from characters to integers

```python
stoi = {ch:i for i, ch in enumerate(chars)}
itos = {i:ch for i, ch in enumerate(chars)}
```

```python
def encode(s):
  return [stoi[c] for c in s]

def decode(l):
  return ''.join([itos[i] for i in l])
```

```python
print(encode("Hello world !"))
```

```python
print(decode([41, 68, 75, 75, 78, 1, 86, 78, 81, 75, 67, 1, 2]))
```

You can use other tokenizer such as Tiktoken (by OpenAI) or SentencePiece (by Google) etc...

Now we will tokenize our entire dataset and separate it between train/val

```python
import torch
```

```python
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
```

```python
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]
```

You don't feed the whole text to a transformer, we will pass it some chunks instead

```python
block_size = 8
train_data[:block_size+1]
```

```python
x = train_data[:block_size]
y = train_data[1:block_size+1]
```

```python
for t in range(block_size):
  context = x[:t+1]
  target = y[t]
  print(f'When input is {context}, the target is {target}')
```

We will generalize this concept

```python
torch.manual_seed(1337)
batch_size = 4 # How many independant sequences will we process in parallel
block_size = 8 # Maximum context lenght for predictions
```

```python
def get_batch(split):
  '''
  Generate a small batch of data inputs x and targets y
  '''
  data = train_data if split == 'train' else val_data
  ix = torch.randint(len(data) - block_size, (batch_size,))
  x = torch.stack([data[i:i+block_size] for i in ix])
  y = torch.stack([data[i+1:i+block_size+1] for i in ix])
  return x, y
```

```python
xb, yb = get_batch('train')
```

# Transformer creation
We will implement one of the simplest model possible : Bigram

```python
import torch.nn as nn
from torch.nn import functional as F
import numpy as np
```

```python
class BigramLanguageModel(nn.Module):

  def __init__(self, vocab_size):
    super().__init__()
    self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)

  def forward(self, idx, targets=None):
    logits = self.token_embedding_table(idx) # (B,T,C)

    if targets == None:
      loss = None
    else:
      B, T, C = logits.shape
      logits = logits.view(B*T,C)
      targets = targets.view(B*T)
      loss = F.cross_entropy(logits, targets)
    return logits, loss

  def generate(self, idx, max_new_tokens):
    for _ in range(max_new_tokens):
      logits, loss = self(idx)
      logits = logits[:,-1,:]
      probs = F.softmax(logits, dim=-1)
      idx_next = torch.multinomial(probs, num_samples=1)
      idx = torch.cat((idx, idx_next), dim=1)
    return idx
```

```python
m = BigramLanguageModel(vocab_size)
logits, loss = m(xb, yb)
```

```python
print(logits.shape)
print(loss)
print(np.log(vocab_size))
```

```python
optimizer = torch.optim.AdamW(m.parameters(), lr=1e-3)
```

```python
batch_size = 32
```

```python
for steps in range(1000):
  xb, yb = get_batch('train')
  logits, loss = m(xb, yb)
  optimizer.zero_grad(set_to_none=True)
  loss.backward()
  optimizer.step()
print(loss.item())
```
